```python
#@title Q2 - Part 1 (100%)
###
def matrix_maker(seed, dims):
    if not isinstance(seed, int):
        print("seed is not integer number.")
        return
    if not isinstance(dims, tuple) or len(dims) != 2:
        print("dims is not valid tuple of size 2.")
        return

    row, col = dims
    matrix = [[0] * col for _ in range(row)]

    for i in range(row ):
        for j in range(col):
            if i == 0 and j == 0:
                matrix[i][j] = seed
            elif j == 0:
                matrix[i][j] = -matrix[i-1][j]
            elif i == 0:
                matrix[i][j] = matrix[i][j-1]
            else:
                matrix[i][j] = matrix[i][j-1] - (matrix[i-1][j-1] + matrix[i-1][j])

    print(matrix)

matrix_maker(1, (3,4))
###
```

# Question 3: (30%)

```python
#@title Q3 - Part 1 (0%)
###
std_num = 9923031
cycle_size = 6
###
```

```python
#@title Q3 - Part 2 (40%)
###
def create_cycle(r: int):
    if r < 3:
        print("radius is less than 3!")
        return

    n = 2 * r - 1
    cycle = np.zeros((n, n), dtype=np.uint8)

    core_i, core_j = n//2, n//2

    for i in range(n):
        for j in range(n):
            distance = np.power((i - core_i), 2)+ np.power((j - core_j), 2)
            if distance <= np.power(r-1, 2):
                cycle[i][j] = 255
            else:
                cycle[i][j] = 0

    return cycle

cycle = create_cycle(cycle_size)
print("Cycle\n", cycle)

###
```

```python
#@title Q3 - Part 3 (35%)
###
def noisy_cycle(cycle, noise_domain):
    if noise_domain < 0:
        print("noise domin must be positive.")
        return


    noisy_cycle = [
            [
                cycle[i][j] + np.random.uniform(low=0, high=noise_domain) if cycle[i][j] == 0
                else cycle[i][j] - np.random.uniform(low=0, high=noise_domain) for j in range(len(cycle[0]))
            ]
            for i in range(len(cycle))
        ]
    noisy_cycle = np.array(noisy_cycle, dtype=np.uint8)
    noisy_cycle = np.floor(noisy_cycle)

    return noisy_cycle

sum_digits = lambda number: 0 if number == 0 else (number % 10) + sum_digits (number // 10)
domain = 20 + sum_digits(std_num) % 15
noisy_cycle = noisy_cycle(cycle=cycle, noise_domain=domain)

print("\nNoisy Cycle\n", noisy_cycle)
###
```

```python
#@title Q3 - Part 4 (15%)
###
fig, axs = plt.subplots(nrows=1, ncols=2)
fig.suptitle(f"HW0-Image-{std_num}")

axs[0].imshow(cycle, cmap='gray')
axs[0].set_title('Gray Image')

axs[1].imshow(noisy_cycle, cmap='gray')
axs[1].set_title(f'Noisy Gray Image ({domain})')
###
```

```python
#@title Q3 - Part 5 (10%)
###
fig = plt.figure(figsize=plt.figaspect(0.5))
fig.suptitle(f"HW0-Surface-{std_num}")

cycle_ax = fig.add_subplot(1, 2, 1, projection='3d')
X, Y = np.meshgrid(np.arange(cycle.shape[0]), np.arange(cycle.shape[1]))
Z = cycle.flatten()
cycle_ax.set_title("Origin Image")
cycle_ax.scatter(X, Y, Z, c='b', marker = '*')



noisy_cycle_ax = fig.add_subplot(1, 2, 2, projection='3d')
X, Y = np.meshgrid(np.arange(noisy_cycle.shape[0]), np.arange(noisy_cycle.shape[1]))
Z = noisy_cycle.flatten()
noisy_cycle_ax.set_title("Noisy Image")
noisy_cycle_ax.scatter(X, Y, Z, c='r', marker = '*')

plt.show()
###
```
